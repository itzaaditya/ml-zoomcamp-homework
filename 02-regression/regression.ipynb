```python
categories_var = ['make','engine_fuel_type','transmission_type','driven_wheels',
'market_category','vehicle_size','vehicle_style']
```

```python
categories = {}

for c in categories_var:
    categories[c] = list(df[c].value_counts().head().index)
```

```python
categories
```

```python
def prepare_X(df):
    df = df.copy()
    features = base.copy()    

    df['age'] = 2017 - df.year
    features.append('age')
    
    for v in [2,3,4]:
        df['num_door_%s' %v] = (df.number_of_doors == v).astype('int')
        features.append('num_door_%s' %v)

    for c,values in categories.items():
        for v in values:
            df['%s_%s' % (c,v)] = (df[c] == v).astype('int')
            features.append('%s_%s' % (c,v))
    
    df_num = df[features]
    df_num = df_num.fillna(0)
    X = df_num.values
    return X
```

```python
X_train = prepare_X(df_train)
w0, w = train_regression(X_train, y_train)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)
rmse(y_val, y_pred)
```

```python
def train_regression_reg(x, y, r=0.001):
    ones = np.ones(x.shape[0])
    x = np.column_stack([ones, x])

    XTX = x.T.dot(x)
    XTX = XTX + r * np.eye(XTX.shape[0])
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(x.T).dot(y)

    return w_full[0], w_full[1:]
    
```

```python
X_train = prepare_X(df_train)
w0, w = train_regression_reg(X_train, y_train, r=0.01)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)
rmse(y_val, y_pred)
```

```python
for i in [0.0, 0.0000001, 0.001, 0.0001, 0.1, 1, 10]:
    X_train = prepare_X(df_train)
    w0, w = train_regression_reg(X_train, y_train, r=i)
    
    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)
    score=rmse(y_val, y_pred)
    print(i,w0,score)


```

```python
r = 0.001
X_train = prepare_X(df_train)
w0, w = train_regression_reg(X_train, y_train, r=r)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)
score = rmse(y_val, y_pred)
score
```
